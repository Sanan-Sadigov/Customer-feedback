```python
import pandas as pd
import numpy as np

customer = pd.read_excel(r'c:\Users\Senan\Desktop\customerfeedback.xlsx', sheet_name=None)
customer.keys()
```

This code merges multiple DataFrames from a `customer` dictionary to create a comprehensive `Customer_feedback_df` DataFrame.

```python
customer_table=customer['Customer Table']
device=customer['Device']
supscription=customer['Subscription']
geography=customer['Geography']
role=customer['Role']

merged_df=pd.merge(customer_table,device, on='Customer ID', how='left')
second_merged_df=pd.merge(merged_df,supscription, on='Subscription ID', how='left')
third_merged_df=pd.merge(second_merged_df,geography, on='Geography ID', how='left')
Customer_feedback_df=pd.merge(third_merged_df,role, on='Role ID', how='left')
Customer_feedback_df
```

This code drops all rows with missing values from the `Customer_feedback_df` DataFrame and then displays the resulting DataFrame.

```python
Customer_feedback_df.dropna(inplace=True)
Customer_feedback_df
```

This code converts the 'Tenure' column in the `Customer_feedback_df` DataFrame to the integer data type.

```python
Customer_feedback_df['Tenure']=Customer_feedback_df['Tenure'].astype('int64')
Customer_feedback_df['Tenure']
```

This code filters and analyzes the `Customer_feedback_df` DataFrame to find the most frequently rated theme in each country with an original score of 5 or higher.(5 or higher is shown as High)

```python
High_score=Customer_feedback_df[Customer_feedback_df['Original Score']>=5]
high_rating_counts = High_score.groupby(['Country-Region', 'Theme']).size().reset_index(name='high_rating_count')

most_used_theme = high_rating_counts.loc[high_rating_counts.groupby('Country-Region')['high_rating_count'].idxmax()].reset_index(drop=True)
most_used_theme
```

This code filters and analyzes the `Customer_feedback_df` DataFrame to find the low frequently rated theme in each country with an original score of less than 5. (lower than 5 is shown as low)

```python
Low_score=Customer_feedback_df[Customer_feedback_df['Original Score']<5]
Low_rating_counts = Low_score.groupby(['Country-Region', 'Theme']).size().reset_index(name='low_rating_count')
Low_rating_counts

less_used_theme = Low_rating_counts.loc[Low_rating_counts.groupby('Country-Region')['low_rating_count'].idxmax()].reset_index(drop=True)
less_used_theme
```

This code groups the `Customer_feedback_df` DataFrame by country and device type, then counts the number of customers for each combination.

```python
used_device = Customer_feedback_df.groupby(['Country-Region','device' ])['Customer ID'].size().reset_index(name='Count_of_devices')
used_device
```

This code groups the `Customer_feedback_df` DataFrame by country and subscription type, then counts the number of customers for each combination.

```python
Subscriptions = Customer_feedback_df.groupby(['Country-Region','Subscription Type' ])['Customer ID'].size().reset_index(name='Count_of_Subscriptions')
Subscriptions
```

This code shows how many customer have for each country.

```python
Customer_feedback_df.groupby('Country-Region')['Customer ID'].count()
```

This code writes multiple DataFrames to different sheets in an Excel file using the `xlsxwriter` engine.

```python
with pd.ExcelWriter(r'c:\Users\Senan\Desktop\analysed_data.xlsx', engine='xlsxwriter') as writer:
    most_used_theme.to_excel(writer, sheet_name='most_used_theme', index=False)
    less_used_theme.to_excel(writer, sheet_name='less_used_theme', index=False)
    Subscriptions.to_excel(writer, sheet_name='Subscriptions', index=False)
    used_device.to_excel(writer, sheet_name='used_device', index=False)
```
